```python
# --------------------------------------------------------
# Display charts of some technical indicators
# -------------
def display_charts(dataName, ti):
    plt.figure(figsize=(9, 7))
    plt.suptitle(dataName, fontsize=14)
    plt.rcParams.update({'font.size': 7})
    plt.tight_layout()
    n = 200

    sp1 = plt.subplot(1, 2, 1)
    sp1.grid()
    sp1.title.set_text('MACD')
    plt.plot(ti['MACD'].iloc[-n:])

    sp2 = plt.subplot(1, 2, 2)
    sp2.grid()
    sp2.title.set_text('RSI14 OBV')
    plt.plot(ti[['RSI', 'OBV']].iloc[-n:])

    plt.show()
```

### Step 2. Try using matplotlib to create 2-level chart (up and down)

**This is example of creating 2 subcharts in one chart using matplotlib!
<font color='red'>Try to modify it your for yourself</font>** and then put the correct data into the appropriate chart.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)

# Some generic data to fill our chart:
x = np.linspace(1,10,50)
y1 = 1/np.sqrt(x)
y2 = np.random.random(50)

# Create our figure and specify the size:
fig = plt.figure(figsize=(12, 5))

# Create the subplots and assign them to a variable name:
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

# Now plot the data on each of the subplots
ax1.plot(x, y1, color='red', label=('$y = 1/\sqrt{x}$'))
ax2.plot(x, y2, linestyle='-.')

# Add a title to each subplot
ax1.set_title('Random Data')
ax2.set_title('More Random Data')

# Add labels to the horizontal and vertical axes
ax1.set_ylabel('Left label')
ax2.set_xlabel('Bottom label')

# This is how we make the legends visible in the subplots:
ax1.legend()
ax2.legend()

plt.show()
```
